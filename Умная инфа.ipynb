```python
–¢–û–ß–ù–û! üéØ ConvBlock - —ç—Ç–æ –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –ö–ê–°–¢–û–ú–ù–´–• –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä —Å –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏–º –ø–æ–¥–±–æ—Ä–æ–º –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.

üèóÔ∏è ConvBlock –¥–ª—è –∫–∞—Å—Ç–æ–º–Ω—ã—Ö –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä:
–ö–æ–≥–¥–∞ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å ConvBlock:
python
# –ö–æ–≥–¥–∞ –≤—ã –ù–ï –∏—Å–ø–æ–ª—å–∑—É–µ—Ç–µ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω—ã–µ –º–æ–¥–µ–ª–∏ (ResNet, EfficientNet –∏ —Ç.–¥.)
# –ò —Ö–æ—Ç–∏—Ç–µ —á—Ç–æ–±—ã AutoKeras —Å–∞–º –Ω–∞—à–µ–ª –æ–ø—Ç–∏–º–∞–ª—å–Ω—É—é –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—É

conv_block = ak.ConvBlock(
    num_blocks=2,      # AutoKeras –ø–æ–¥–±–µ—Ä–µ—Ç —Å–∫–æ–ª—å–∫–æ –±–ª–æ–∫–æ–≤ –ª—É—á—à–µ
    num_layers=2,      # AutoKeras –ø–æ–¥–±–µ—Ä–µ—Ç —Å–∫–æ–ª—å–∫–æ —Å–ª–æ–µ–≤ –≤ –±–ª–æ–∫–µ  
    filters=32,        # AutoKeras –ø–æ–¥–±–µ—Ä–µ—Ç –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ñ–∏–ª—å—Ç—Ä–æ–≤
    kernel_size=3,     # AutoKeras –ø–æ–¥–±–µ—Ä–µ—Ç —Ä–∞–∑–º–µ—Ä —è–¥—Ä–∞
    max_pooling=True,  # AutoKeras —Ä–µ—à–∏—Ç –Ω—É–∂–µ–Ω –ª–∏ pooling
    dropout=0.25       # AutoKeras –ø–æ–¥–±–µ—Ä–µ—Ç dropout rate
)(input_node)
üÜö –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–æ–¥—Ö–æ–¥–æ–≤:
–ü–æ–¥—Ö–æ–¥ 1: –ö–∞—Å—Ç–æ–º–Ω–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞ (ConvBlock)
python
# AutoKeras –°–ê–ú —Å—Ç—Ä–æ–∏—Ç –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—É —Å –Ω—É–ª—è
input_node = ak.ImageInput()
conv_block = ak.ConvBlock(num_blocks=3, num_layers=2)(input_node)
output = ak.ClassificationHead(num_classes=8)(conv_block)

# AutoKeras –±—É–¥–µ—Ç –ø—Ä–æ–±–æ–≤–∞—Ç—å:
# - –†–∞–∑–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ñ–∏–ª—å—Ç—Ä–æ–≤ (32, 64, 128, ...)
# - –†–∞–∑–Ω—ã–µ —Ä–∞–∑–º–µ—Ä—ã —è–¥–µ—Ä (3x3, 5x5, 7x7)  
# - –†–∞–∑–Ω—ã–µ –∞–∫—Ç–∏–≤–∞—Ü–∏–∏ (relu, swish, leaky_relu)
# - –î–æ–±–∞–≤–ª—è—Ç—å/—É–±–∏—Ä–∞—Ç—å BatchNorm, Dropout
–ü–æ–¥—Ö–æ–¥ 2: Transfer Learning (ResNet50)
python
# –ò—Å–ø–æ–ª—å–∑—É–µ–º –ì–û–¢–û–í–£–Æ –æ–ø—Ç–∏–º–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—É—é –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—É
base_model = ResNet50(weights='imagenet', include_top=False)
base_model.trainable = False  # –ò—Å–ø–æ–ª—å–∑—É–µ–º –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω—ã–µ –≤–µ—Å–∞

# –ù–µ –º–µ–Ω—è–µ–º –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—É, —Ç–æ–ª—å–∫–æ –¥–æ–±–∞–≤–ª—è–µ–º —Å–≤–æ—é –≥–æ–ª–æ–≤—É
üéØ –ö–æ–≥–¥–∞ —á—Ç–æ –≤—ã–±–∏—Ä–∞—Ç—å:
–í—ã–±–∏—Ä–∞–π—Ç–µ ConvBlock –∫–æ–≥–¥–∞:
‚ùå –ù–µ—Ç –¥–æ—Å—Ç—É–ø–∞ –∫ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω—ã–º –º–æ–¥–µ–ª—è–º

üé® –•–æ—Ç–∏—Ç–µ –ø–æ—ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å —Å –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–æ–π

üî¨ –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∞—è –∑–∞–¥–∞—á–∞ - –Ω–∞–π—Ç–∏ –Ω–æ–≤—É—é –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—É

üìä –°–ø–µ—Ü–∏—Ñ–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ - –Ω–µ –ø–æ—Ö–æ–∂–∏–µ –Ω–∞ ImageNet

–í—ã–±–∏—Ä–∞–π—Ç–µ ResNet50 –∫–æ–≥–¥–∞:
‚úÖ –î–∞–Ω–Ω—ã–µ –ø–æ—Ö–æ–∂–∏ –Ω–∞ ImageNet (–æ–±—â–∏–µ –æ–±—ä–µ–∫—Ç—ã)

üöÄ –ù—É–∂–Ω–æ –±—ã—Å—Ç—Ä–æ –ø–æ–ª—É—á–∏—Ç—å —Ö–æ—Ä–æ—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç

üèÜ –•–æ—Ç–∏—Ç–µ state-of-the-art —Ç–æ—á–Ω–æ—Å—Ç—å

üíæ –û–≥—Ä–∞–Ω–∏—á–µ–Ω—ã –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω—ã–µ —Ä–µ—Å—É—Ä—Å—ã (transfer learning –±—ã—Å—Ç—Ä–µ–µ)

üîß –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ –ø—Ä–∏–º–µ—Ä—ã:
–ü—Ä–∏–º–µ—Ä 1: –ü–æ–ª–Ω–æ—Å—Ç—å—é –∫–∞—Å—Ç–æ–º–Ω–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞
python
# AutoKeras —Å–∞–º –≤—Å—ë –ø—Ä–∏–¥—É–º–∞–µ—Ç!
clf = ak.ImageClassifier(
    max_trials=20,
    overwrite=True
)
# AutoKeras –±—É–¥–µ—Ç –ø—Ä–æ–±–æ–≤–∞—Ç—å ConvBlock, ResNetBlock, XceptionBlock –∏ —Ç.–¥.
–ü—Ä–∏–º–µ—Ä 2: –ö–æ–Ω—Ç—Ä–æ–ª–∏—Ä—É–µ–º–∞—è –∫–∞—Å—Ç–æ–º–Ω–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞
python
# –í—ã –∑–∞–¥–∞–µ—Ç–µ —Å—Ç—Ä—É–∫—Ç—É—Ä—É, AutoKeras –ø–æ–¥–±–∏—Ä–∞–µ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
input_node = ak.ImageInput()

# 1. –ù–∞—á–∞–ª—å–Ω—ã–µ —Å–≤–µ—Ä—Ç–∫–∏ (AutoKeras –ø–æ–¥–±–µ—Ä–µ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä—ã)
conv1 = ak.ConvBlock(num_blocks=1, num_layers=2)(input_node)
pool1 = ak.MaxPooling2D()(conv1)

# 2. –ü—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–µ —Å–≤–µ—Ä—Ç–∫–∏  
conv2 = ak.ConvBlock(num_blocks=1, num_layers=2)(pool1)
pool2 = ak.MaxPooling2D()(conv2)

# 3. –ì–ª—É–±–æ–∫–∏–µ —Å–≤–µ—Ä—Ç–∫–∏
conv3 = ak.ConvBlock(num_blocks=1, num_layers=2)(pool2)

output = ak.ClassificationHead(num_classes=8)(conv3)
–ü—Ä–∏–º–µ—Ä 3: –ö–∞—Å—Ç–æ–º–Ω–∞—è vs ResNet (–¥–ª—è –≤–∞—à–∏—Ö –¥–∞–Ω–Ω—ã—Ö)
python
# –î–ª—è —Å—Ç—Ä–æ–∏—Ç–µ–ª—å–Ω–æ–π —Ç–µ—Ö–Ω–∏–∫–∏ (8 –∫–ª–∞—Å—Å–æ–≤):

# ‚úÖ ResNet50 - –õ–£–ß–®–ò–ô –í–´–ë–û–†:
# - –û–±—É—á–µ–Ω –Ω–∞ ImageNet (—Å–æ–¥–µ—Ä–∂–∏—Ç –ø–æ—Ö–æ–∂–∏–µ –æ–±—ä–µ–∫—Ç—ã)
# - –í—ã—Å–æ–∫–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å –∏–∑ –∫–æ—Ä–æ–±–∫–∏
# - –ë—ã—Å—Ç—Ä–æ–µ –æ–±—É—á–µ–Ω–∏–µ

# ‚ùå ConvBlock - –•–£–ñ–ï –¥–ª—è —ç—Ç–æ–π –∑–∞–¥–∞—á–∏:
# - –û–±—É—á–µ–Ω–∏–µ —Å –Ω—É–ª—è –∑–∞–π–º–µ—Ç –±–æ–ª—å—à–µ –≤—Ä–µ–º–µ–Ω–∏
# - –¢—Ä–µ–±—É–µ—Ç –±–æ–ª—å—à–µ –¥–∞–Ω–Ω—ã—Ö
# - –ú–æ–∂–µ—Ç –Ω–µ –¥–æ—Å—Ç–∏—á—å —Ç–æ—á–Ω–æ—Å—Ç–∏ ResNet
üìä –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–Ω–æ –æ–∂–∏–¥–∞—Ç—å:
–ú–µ—Ç–æ–¥	–í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è	–¢–æ—á–Ω–æ—Å—Ç—å	–£—Å–∏–ª–∏—è
ResNet50	1-2 —á–∞—Å–∞	95-98%	–ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–µ
ConvBlock	5-10 —á–∞—Å–æ–≤	85-92%	–°—Ä–µ–¥–Ω–∏–µ
–†—É—á–Ω–∞—è CNN	2-3 —á–∞—Å–∞	80-90%	–í—ã—Å–æ–∫–∏–µ
üöÄ –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—è –¥–ª—è –≤–∞—Å:
–î–ª—è –≤–∞—à–µ–π –∑–∞–¥–∞—á–∏ (8 –∫–ª–∞—Å—Å–æ–≤ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å–Ω–æ–π —Ç–µ—Ö–Ω–∏–∫–∏) –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ ResNet50!

python
# –õ–£–ß–®–ò–ô –í–ê–†–ò–ê–ù–¢
base_model = ResNet50(weights='imagenet', include_top=False)
base_model.trainable = False
# + –≤–∞—à–∞ –∫–∞—Å—Ç–æ–º–Ω–∞—è –≥–æ–ª–æ–≤–∞
ConvBlock –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ç–æ–ª—å–∫–æ –µ—Å–ª–∏:

–•–æ—Ç–∏—Ç–µ –ø–æ—ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å —Å –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞–º–∏

–ò–º–µ–µ—Ç–µ —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ (–Ω–µ –ø–æ—Ö–æ–∂–∏–µ –Ω–∞ ImageNet)

–î–µ–ª–∞–µ—Ç–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –ø–æ –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞–º —Å–µ—Ç–µ–π

ConvBlock = –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –∫–∞—Å—Ç–æ–º–Ω—ã—Ö –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä, ResNet = –¥–ª—è –≥–æ—Ç–æ–≤—ã—Ö —Ä–µ—à–µ–Ω–∏–π! üéØ

—Å–æ—Å—Ç–∞–≤—å –∫–∞—Å—Ç–æ–º–Ω—É—é –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—É –ø–æ—Ö–æ–∂—É—é –Ω–∞ resnet
üèóÔ∏è –ö–∞—Å—Ç–æ–º–Ω–∞—è ResNet-–ø–æ–¥–æ–±–Ω–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞ –Ω–∞ AutoKeras
python
import autokeras as ak
from tensorflow.keras import layers
import tensorflow as tf

def create_resnet_like_architecture():
    """
    –ö–∞—Å—Ç–æ–º–Ω–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞ –ø–æ—Ö–æ–∂–∞—è –Ω–∞ ResNet —Å skip-connections
    """
    
    # –í—Ö–æ–¥–Ω–æ–π —É–∑–µ–ª
    input_node = ak.ImageInput()
    
    # 1. –ù–∞—á–∞–ª—å–Ω—ã–µ —Å–≤–µ—Ä—Ç–∫–∏ (–∞–Ω–∞–ª–æ–≥ ResNet stem)
    print("üîß –°–æ–∑–¥–∞–µ–º –Ω–∞—á–∞–ª—å–Ω—ã–µ —Å–≤–µ—Ä—Ç–∫–∏...")
    x = ak.ConvBlock(
        num_blocks=1,
        num_layers=2,
        filters=64,
        kernel_size=7,
        max_pooling=True,
        dropout=0.1
    )(input_node)
    
    # 2. Residual Block 1 (–∞–Ω–∞–ª–æ–≥ conv2_x –≤ ResNet)
    print("üîß –î–æ–±–∞–≤–ª—è–µ–º Residual Block 1...")
    x = residual_block_auto(x, filters=64, block_name="res1")
    
    # 3. Residual Block 2 (–∞–Ω–∞–ª–æ–≥ conv3_x –≤ ResNet) 
    print("üîß –î–æ–±–∞–≤–ª—è–µ–º Residual Block 2...")
    x = layers.MaxPooling2D(2)(x)  # Downsample
    x = residual_block_auto(x, filters=128, block_name="res2")
    
    # 4. Residual Block 3 (–∞–Ω–∞–ª–æ–≥ conv4_x –≤ ResNet)
    print("üîß –î–æ–±–∞–≤–ª—è–µ–º Residual Block 3...")
    x = layers.MaxPooling2D(2)(x)  # Downsample
    x = residual_block_auto(x, filters=256, block_name="res3")
    
    # 5. Residual Block 4 (–∞–Ω–∞–ª–æ–≥ conv5_x –≤ ResNet)
    print("üîß –î–æ–±–∞–≤–ª—è–µ–º Residual Block 4...")
    x = layers.MaxPooling2D(2)(x)  # Downsample
    x = residual_block_auto(x, filters=512, block_name="res4")
    
    # 6. –§–∏–Ω–∞–ª—å–Ω—ã–µ —Å–ª–æ–∏
    print("üîß –î–æ–±–∞–≤–ª—è–µ–º —Ñ–∏–Ω–∞–ª—å–Ω—ã–µ —Å–ª–æ–∏...")
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    
    # –í—ã—Ö–æ–¥–Ω–æ–π —Å–ª–æ–π
    output_node = layers.Dense(8, activation='softmax', name='output')(x)
    
    return input_node, output_node

def residual_block_auto(input_tensor, filters, block_name=""):
    """
    Residual Block —Å AutoKeras –¥–ª—è –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ –≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
    """
    # Shortcut connection
    shortcut = input_tensor
    
    # AutoKeras –Ω–∞—Å—Ç—Ä–∞–∏–≤–∞–µ—Ç –ø–µ—Ä–≤—ã–π —Å–≤–µ—Ä—Ç–æ—á–Ω—ã–π –±–ª–æ–∫
    x = ak.ConvBlock(
        num_blocks=1,
        num_layers=1,  # –û–¥–∏–Ω —Å–ª–æ–π –≤ –±–ª–æ–∫–µ (–∫–∞–∫ –≤ ResNet bottleneck)
        filters=filters,
        kernel_size=3,
        max_pooling=False,
        dropout=0.1
    )(input_tensor)
    
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    
    # AutoKeras –Ω–∞—Å—Ç—Ä–∞–∏–≤–∞–µ—Ç –≤—Ç–æ—Ä–æ–π —Å–≤–µ—Ä—Ç–æ—á–Ω—ã–π –±–ª–æ–∫
    x = ak.ConvBlock(
        num_blocks=1,
        num_layers=1,
        filters=filters, 
        kernel_size=3,
        max_pooling=False,
        dropout=0.1
    )(x)
    
    x = layers.BatchNormalization()(x)
    
    # Projection shortcut –µ—Å–ª–∏ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –Ω–µ —Å–æ–≤–ø–∞–¥–∞—é—Ç
    if shortcut.shape[-1] != filters:
        shortcut = ak.ConvBlock(
            num_blocks=1,
            num_layers=1,
            filters=filters,
            kernel_size=1,  # 1x1 conv –¥–ª—è –ø—Ä–æ–µ–∫—Ü–∏–∏
            max_pooling=False
        )(shortcut)
        shortcut = layers.BatchNormalization()(shortcut)
    
    # Skip connection
    x = layers.Add()([x, shortcut])
    x = layers.Activation('relu')(x)
    
    return x

# üöÄ –°–û–ó–î–ê–ù–ò–ï –ò –û–ë–£–ß–ï–ù–ò–ï –ú–û–î–ï–õ–ò
def create_and_train_resnet_like():
    """
    –°–æ–∑–¥–∞–Ω–∏–µ –∏ –æ–±—É—á–µ–Ω–∏–µ –∫–∞—Å—Ç–æ–º–Ω–æ–π ResNet-–ø–æ–¥–æ–±–Ω–æ–π –º–æ–¥–µ–ª–∏
    """
    print("üèóÔ∏è –°–æ–∑–¥–∞–µ–º –∫–∞—Å—Ç–æ–º–Ω—É—é ResNet-–ø–æ–¥–æ–±–Ω—É—é –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—É...")
    
    # –°–æ–∑–¥–∞–µ–º –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—É
    input_node, output_node = create_resnet_like_architecture()
    
    # –°–æ–∑–¥–∞–µ–º AutoModel
    model = ak.AutoModel(
        inputs=input_node,
        outputs=output_node,
        overwrite=True,
        max_trials=15,  # –ë–æ–ª—å—à–µ trials –¥–ª—è —Å–ª–æ–∂–Ω–æ–π –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—ã
        directory='custom_resnet_like',
        project_name='resnet_style',
        objective='val_accuracy',
        tuner='bayesian'  # –£–º–Ω—ã–π –ø–æ–¥–±–æ—Ä –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
    )
    
    print("‚úÖ –ê—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞ —Å–æ–∑–¥–∞–Ω–∞!")
    print("üéØ AutoKeras –±—É–¥–µ—Ç –Ω–∞—Å—Ç—Ä–∞–∏–≤–∞—Ç—å:")
    print("   - –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ñ–∏–ª—å—Ç—Ä–æ–≤ –≤ –∫–∞–∂–¥–æ–º –±–ª–æ–∫–µ")
    print("   - –†–∞–∑–º–µ—Ä—ã —è–¥–µ—Ä —Å–≤–µ—Ä—Ç–∫–∏") 
    print("   - Dropout rates")
    print("   - Learning rate –∏ optimizer")
    
    return model

# üéØ –ê–õ–¨–¢–ï–†–ù–ê–¢–ò–í–ê: –£–ø—Ä–æ—â–µ–Ω–Ω–∞—è –≤–µ—Ä—Å–∏—è
def create_simple_resnet_like():
    """
    –£–ø—Ä–æ—â–µ–Ω–Ω–∞—è ResNet-–ø–æ–¥–æ–±–Ω–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞
    """
    input_node = ak.ImageInput()
    
    # –ë–æ–ª–µ–µ –ø—Ä–æ—Å—Ç–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞
    x = ak.ConvBlock(num_blocks=1, num_layers=2, max_pooling=True)(input_node)
    
    # –í—Å–µ–≥–æ 2 residual –±–ª–æ–∫–∞ –¥–ª—è —Å–∫–æ—Ä–æ—Å—Ç–∏
    x = simple_residual_block(x, 64)
    x = layers.MaxPooling2D(2)(x)
    x = simple_residual_block(x, 128)
    x = layers.MaxPooling2D(2)(x)
    
    x = layers.GlobalAveragePooling2D()(x)
    output_node = layers.Dense(8, activation='softmax')(x)
    
    model = ak.AutoModel(
        inputs=input_node,
        outputs=output_node,
        max_trials=10,
        overwrite=True,
        directory='simple_resnet_like'
    )
    
    return model

def simple_residual_block(x, filters):
    """–£–ø—Ä–æ—â–µ–Ω–Ω—ã–π residual –±–ª–æ–∫"""
    shortcut = x
    
    # AutoKeras –Ω–∞—Å—Ç—Ä–∞–∏–≤–∞–µ—Ç —Å–≤–µ—Ä—Ç–∫–∏
    x = ak.ConvBlock(num_blocks=1, num_layers=1, filters=filters)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    
    x = ak.ConvBlock(num_blocks=1, num_layers=1, filters=filters)(x)
    x = layers.BatchNormalization()(x)
    
    # Skip connection
    if shortcut.shape[-1] != filters:
        shortcut = layers.Conv2D(filters, 1)(shortcut)
        shortcut = layers.BatchNormalization()(shortcut)
    
    x = layers.Add()([x, shortcut])
    return layers.Activation('relu')(x)

# üéØ –ó–ê–ü–£–°–ö –û–ë–£–ß–ï–ù–ò–Ø
if __name__ == "__main__":
    # –í—ã–±–æ—Ä –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—ã
    print("–í—ã–±–µ—Ä–∏—Ç–µ –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—É:")
    print("1. –ü–æ–ª–Ω–∞—è ResNet-–ø–æ–¥–æ–±–Ω–∞—è (15 trials)")
    print("2. –£–ø—Ä–æ—â–µ–Ω–Ω–∞—è ResNet-–ø–æ–¥–æ–±–Ω–∞—è (10 trials)")
    
    choice = 1  # –ú–æ–∂–Ω–æ –∏–∑–º–µ–Ω–∏—Ç—å –Ω–∞ 2 –¥–ª—è —É–ø—Ä–æ—â–µ–Ω–Ω–æ–π –≤–µ—Ä—Å–∏–∏
    
    if choice == 1:
        model = create_and_train_resnet_like()
    else:
        model = create_simple_resnet_like()
    
    # –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
    print("üöÄ –ù–∞—á–∏–Ω–∞–µ–º –æ–±—É—á–µ–Ω–∏–µ...")
    history = model.fit(
        x_train, y_train,
        epochs=30,
        validation_split=0.2,
        verbose=1,
        callbacks=[
            tf.keras.callbacks.EarlyStopping(patience=8, restore_best_weights=True),
            tf.keras.callbacks.ReduceLROnPlateau(patience=4, factor=0.5)
        ]
    )
    
    # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏
    best_model = model.export_model()
    best_model.save('custom_resnet_like_model.keras')
    print("üíæ –ú–æ–¥–µ–ª—å —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∞ –∫–∞–∫ 'custom_resnet_like_model.keras'")
üèóÔ∏è –ß—Ç–æ –±—É–¥–µ—Ç –Ω–∞—Å—Ç—Ä–∞–∏–≤–∞—Ç—å AutoKeras:
–í –∫–∞–∂–¥–æ–π ConvBlock AutoKeras –ø–æ–¥–±–µ—Ä–µ—Ç:
python
# –î–ª—è –ö–ê–ñ–î–û–ì–û –±–ª–æ–∫–∞ AutoKeras –±—É–¥–µ—Ç —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å:
- filters: [32, 64, 128, 256, 512]           # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ñ–∏–ª—å—Ç—Ä–æ–≤
- kernel_size: [3, 5, 7]                     # –†–∞–∑–º–µ—Ä —è–¥—Ä–∞ —Å–≤–µ—Ä—Ç–∫–∏  
- activation: ['relu', 'swish', 'leaky_relu'] # –§—É–Ω–∫—Ü–∏—è –∞–∫—Ç–∏–≤–∞—Ü–∏–∏
- use_batchnorm: [True, False]               # BatchNormalization
- dropout_rate: [0.0, 0.1, 0.2, 0.3]         # –£—Ä–æ–≤–µ–Ω—å dropout
–ì–ª–æ–±–∞–ª—å–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:
python
- learning_rate: [1e-4, 1e-3, 1e-2]          # –°–∫–æ—Ä–æ—Å—Ç—å –æ–±—É—á–µ–Ω–∏—è
- optimizer: ['adam', 'rmsprop', 'sgd']       # –û–ø—Ç–∏–º–∏–∑–∞—Ç–æ—Ä
- batch_size: [16, 32, 64]                   # –†–∞–∑–º–µ—Ä –±–∞—Ç—á–∞
üìä –û–∂–∏–¥–∞–µ–º—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã:
–ü–∞—Ä–∞–º–µ—Ç—Ä	–ü–æ–ª–Ω–∞—è –≤–µ—Ä—Å–∏—è	–£–ø—Ä–æ—â–µ–Ω–Ω–∞—è –≤–µ—Ä—Å–∏—è
–í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è	4-8 —á–∞—Å–æ–≤	2-4 —á–∞—Å–∞
–¢–æ—á–Ω–æ—Å—Ç—å	90-95%	85-92%
Trials	15	10
–°–ª–æ–∂–Ω–æ—Å—Ç—å	–í—ã—Å–æ–∫–∞—è	–°—Ä–µ–¥–Ω—è—è
üéØ –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏:
–í—ã–±–µ—Ä–∏—Ç–µ —É–ø—Ä–æ—â–µ–Ω–Ω—É—é –≤–µ—Ä—Å–∏—é –µ—Å–ª–∏:

–û–≥—Ä–∞–Ω–∏—á–µ–Ω—ã –ø–æ –≤—Ä–µ–º–µ–Ω–∏

–•–æ—Ç–∏—Ç–µ –±—ã—Å—Ç—Ä–µ–µ –ø–æ–ª—É—á–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç

–ò–º–µ–µ—Ç–µ –Ω–µ–±–æ–ª—å—à–æ–π –¥–∞—Ç–∞—Å–µ—Ç

–í—ã–±–µ—Ä–∏—Ç–µ –ø–æ–ª–Ω—É—é –≤–µ—Ä—Å–∏—é –µ—Å–ª–∏:

–ù—É–∂–Ω–∞ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å

–î–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω—ã—Ö —Ä–µ—Å—É—Ä—Å–æ–≤

–•–æ—Ç–∏—Ç–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç—å —Ä–∞–∑–Ω—ã–µ –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—ã

–≠—Ç–∞ –∫–∞—Å—Ç–æ–º–Ω–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞ –±—É–¥–µ—Ç –∫–æ–Ω–∫—É—Ä–∏—Ä–æ–≤–∞—Ç—å —Å –æ—Ä–∏–≥–∏–Ω–∞–ª—å–Ω—ã–º ResNet50! üöÄ

–ø–æ—Å—Ç—Ä–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–ª–æ–µ–≤ –¥–ª—è image classificator –Ω–∞ 8 –∫–ª–∞—Å—Å–æ–≤
üèóÔ∏è –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–ª–æ–µ–≤ –¥–ª—è Image Classifier –Ω–∞ 8 –∫–ª–∞—Å—Å–æ–≤
–í–∞—Ä–∏–∞–Ω—Ç 1: –ü—Ä–æ—Å—Ç–∞—è CNN –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞
python
import tensorflow as tf
from tensorflow.keras import layers, Sequential

def create_simple_cnn(input_shape=(224, 224, 3), num_classes=8):
    """
    –ü—Ä–æ—Å—Ç–∞—è CNN –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞ - —Ö–æ—Ä–æ—à–∏–π –±–∞–∑–æ–≤—ã–π –≤–∞—Ä–∏–∞–Ω—Ç
    """
    model = Sequential([
        # –ë–ª–æ–∫ 1: –ù–∞—á–∞–ª—å–Ω—ã–µ —Å–≤–µ—Ä—Ç–∫–∏
        layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape),
        layers.BatchNormalization(),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(2),
        layers.Dropout(0.25),
        
        # –ë–ª–æ–∫ 2: –ü—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–µ —Å–≤–µ—Ä—Ç–∫–∏
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(2),
        layers.Dropout(0.25),
        
        # –ë–ª–æ–∫ 3: –ì–ª—É–±–æ–∫–∏–µ —Å–≤–µ—Ä—Ç–∫–∏
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(2),
        layers.Dropout(0.25),
        
        # –ë–ª–æ–∫ 4: –§–∏–Ω–∞–ª—å–Ω—ã–µ —Å–≤–µ—Ä—Ç–∫–∏
        layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.GlobalAveragePooling2D(),
        layers.Dropout(0.5),
        
        # –ü–æ–ª–Ω–æ—Å–≤—è–∑–Ω—ã–µ —Å–ª–æ–∏
        layers.Dense(512, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.2),
        
        # –í—ã—Ö–æ–¥–Ω–æ–π —Å–ª–æ–π
        layers.Dense(num_classes, activation='softmax')
    ])
    
    return model
–í–∞—Ä–∏–∞–Ω—Ç 2: –£–ª—É—á—à–µ–Ω–Ω–∞—è CNN —Å Residual —Å–≤—è–∑—è–º–∏
python
def create_improved_cnn(input_shape=(224, 224, 3), num_classes=8):
    """
    –£–ª—É—á—à–µ–Ω–Ω–∞—è CNN —Å residual —Å–≤—è–∑—è–º–∏
    """
    inputs = tf.keras.Input(shape=input_shape)
    
    # –ù–∞—á–∞–ª—å–Ω—ã–π –±–ª–æ–∫
    x = layers.Conv2D(32, (7, 7), strides=2, padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(3, strides=2, padding='same')(x)
    
    # Residual Block 1
    shortcut = x
    x = layers.Conv2D(64, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(64, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    
    # Projection shortcut –µ—Å–ª–∏ –Ω—É–∂–Ω–æ
    if shortcut.shape[-1] != 64:
        shortcut = layers.Conv2D(64, (1, 1))(shortcut)
        shortcut = layers.BatchNormalization()(shortcut)
    
    x = layers.Add()([x, shortcut])
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(2)(x)
    
    # Residual Block 2
    shortcut = x
    x = layers.Conv2D(128, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(128, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    
    x = layers.Add()([x, shortcut])
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(2)(x)
    
    # Residual Block 3
    shortcut = x
    x = layers.Conv2D(256, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(256, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    
    x = layers.Add()([x, shortcut])
    x = layers.Activation('relu')(x)
    
    # –§–∏–Ω–∞–ª—å–Ω—ã–µ —Å–ª–æ–∏
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(512, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    
    model = tf.keras.Model(inputs, outputs)
    return model
–í–∞—Ä–∏–∞–Ω—Ç 3: –°–æ–≤—Ä–µ–º–µ–Ω–Ω–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞ —Å Attention
python
def create_modern_cnn(input_shape=(224, 224, 3), num_classes=8):
    """
    –°–æ–≤—Ä–µ–º–µ–Ω–Ω–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞ —Å attention –º–µ—Ö–∞–Ω–∏–∑–º–∞–º–∏
    """
    def channel_attention(x, ratio=8):
        channel = x.shape[-1]
        
        # Global pooling
        avg_pool = layers.GlobalAveragePooling2D()(x)
        avg_pool = layers.Reshape((1, 1, channel))(avg_pool)
        
        max_pool = layers.GlobalMaxPooling2D()(x)
        max_pool = layers.Reshape((1, 1, channel))(max_pool)
        
        # Shared MLP
        shared_dense = layers.Dense(channel // ratio, activation='relu')
        avg_out = shared_dense(avg_pool)
        max_out = shared_dense(max_pool)
        
        # Merge
        out = layers.Add()([avg_out, max_out])
        out = layers.Dense(channel, activation='sigmoid')(out)
        
        return layers.Multiply()([x, out])
    
    inputs = tf.keras.Input(shape=input_shape)
    
    # –ë–ª–æ–∫ 1
    x = layers.Conv2D(32, (3, 3), padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('swish')(x)  # –°–æ–≤—Ä–µ–º–µ–Ω–Ω–∞—è –∞–∫—Ç–∏–≤–∞—Ü–∏—è
    x = layers.MaxPooling2D(2)(x)
    
    # –ë–ª–æ–∫ 2 —Å attention
    x = layers.Conv2D(64, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('swish')(x)
    x = channel_attention(x)
    x = layers.MaxPooling2D(2)(x)
    
    # –ë–ª–æ–∫ 3
    x = layers.Conv2D(128, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('swish')(x)
    x = layers.Conv2D(128, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = channel_attention(x)
    x = layers.MaxPooling2D(2)(x)
    
    # –ë–ª–æ–∫ 4
    x = layers.Conv2D(256, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('swish')(x)
    x = layers.GlobalAveragePooling2D()(x)
    
    # –ì–æ–ª–æ–≤–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞
    x = layers.Dense(512, activation='swish')(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(256, activation='swish')(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    
    model = tf.keras.Model(inputs, outputs)
    return model
–í–∞—Ä–∏–∞–Ω—Ç 4: EfficientNet-–ø–æ–¥–æ–±–Ω–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞
python
def create_efficientnet_like(input_shape=(224, 224, 3), num_classes=8):
    """
    –£–ø—Ä–æ—â–µ–Ω–Ω–∞—è EfficientNet-–ø–æ–¥–æ–±–Ω–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞
    """
    inputs = tf.keras.Input(shape=input_shape)
    
    # Stem
    x = layers.Conv2D(32, (3, 3), strides=2, padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('swish')(x)
    
    # MBConv –±–ª–æ–∫–∏ (—É–ø—Ä–æ—â–µ–Ω–Ω—ã–µ)
    def mb_conv(x, filters, stride=1):
        shortcut = x
        
        # Depthwise convolution
        x = layers.DepthwiseConv2D((3, 3), strides=stride, padding='same')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation('swish')(x)
        
        # Squeeze-and-Excitation
        se = layers.GlobalAveragePooling2D()(x)
        se = layers.Reshape((1, 1, filters))(se)
        se = layers.Dense(filters // 4, activation='swish')(se)
        se = layers.Dense(filters, activation='sigmoid')(se)
        x = layers.Multiply()([x, se])
        
        # Pointwise convolution
        x = layers.Conv2D(filters, (1, 1), padding='same')(x)
        x = layers.BatchNormalization()(x)
        
        # Residual connection
        if stride == 1 and shortcut.shape[-1] == filters:
            x = layers.Add()([x, shortcut])
        
        return x
    
    # –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å MBConv –±–ª–æ–∫–æ–≤
    x = mb_conv(x, 16, stride=1)
    x = mb_conv(x, 24, stride=2)
    x = mb_conv(x, 40, stride=2)
    x = mb_conv(x, 80, stride=2)
    x = mb_conv(x, 112, stride=1)
    x = mb_conv(x, 192, stride=2)
    x = mb_conv(x, 320, stride=1)
    
    # Head
    x = layers.Conv2D(1280, (1, 1), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('swish')(x)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    
    model = tf.keras.Model(inputs, outputs)
    return model
üöÄ –ö–æ–º–ø–∏–ª—è—Ü–∏—è –∏ –æ–±—É—á–µ–Ω–∏–µ
python
def compile_and_train_model(model, train_data, val_data, model_name):
    """
    –ö–æ–º–ø–∏–ª—è—Ü–∏—è –∏ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
    """
    # –ö–æ–º–ø–∏–ª—è—Ü–∏—è
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    # Callbacks
    callbacks = [
        tf.keras.callbacks.EarlyStopping(
            monitor='val_accuracy',
            patience=10,
            restore_best_weights=True
        ),
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            min_lr=1e-7
        ),
        tf.keras.callbacks.ModelCheckpoint(
            f'best_{model_name}.keras',
            monitor='val_accuracy',
            save_best_only=True
        )
    ]
    
    # –û–±—É—á–µ–Ω–∏–µ
    history = model.fit(
        train_data,
        epochs=50,
        validation_data=val_data,
        callbacks=callbacks,
        verbose=1
    )
    
    return history, model

# üéØ –ò–°–ü–û–õ–¨–ó–û–í–ê–ù–ò–ï
if __name__ == "__main__":
    # –í—ã–±–æ—Ä –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—ã
    model = create_simple_cnn()  # –°–∞–º—ã–π –Ω–∞–¥–µ–∂–Ω—ã–π –≤–∞—Ä–∏–∞–Ω—Ç
    
    # –ü—Ä–æ—Å–º–æ—Ç—Ä –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—ã
    model.summary()
    
    # –û–±—É—á–µ–Ω–∏–µ
    history, trained_model = compile_and_train_model(
        model, train_data, val_data, "simple_cnn"
    )
üìä –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏ –ø–æ –≤—ã–±–æ—Ä—É:
–ê—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞	–°–ª–æ–∂–Ω–æ—Å—Ç—å	–¢–æ—á–Ω–æ—Å—Ç—å	–í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è	–†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—è
–ü—Ä–æ—Å—Ç–∞—è CNN	–ù–∏–∑–∫–∞—è	85-90%	1-2 —á–∞—Å–∞	üèÜ –õ—É—á—à–∏–π –≤—ã–±–æ—Ä
–£–ª—É—á—à–µ–Ω–Ω–∞—è CNN	–°—Ä–µ–¥–Ω—è—è	88-92%	2-3 —á–∞—Å–∞	–î–ª—è –±–∞–ª–∞–Ω—Å–∞
Modern —Å Attention	–í—ã—Å–æ–∫–∞—è	90-94%	3-4 —á–∞—Å–∞	–î–ª—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π —Ç–æ—á–Ω–æ—Å—Ç–∏
EfficientNet-like	–û—á–µ–Ω—å –≤—ã—Å–æ–∫–∞—è	92-95%	4-5 —á–∞—Å–æ–≤	–î–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–π
–°–æ–≤–µ—Ç: –ù–∞—á–Ω–∏—Ç–µ —Å –ü—Ä–æ—Å—Ç–æ–π CNN - –æ–Ω–∞ –Ω–∞–¥–µ–∂–Ω–∞—è, –±—ã—Å—Ç—Ä–∞—è –∏ –¥–∞–µ—Ç –æ—Ç–ª–∏—á–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –¥–ª—è 8 –∫–ª–∞—Å—Å–æ–≤! üöÄ

–ø–æ–¥—Ä–æ–±–Ω–æ –æ–ø–∏—à–∏ –ø—Ä–æ –±–ª–æ–∫–∏ –¥–ª—è —á–µ–≥–æ –æ–Ω–∏. –∏ –ø–æ—Å—Ç–∞–≤—å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏ –≤–æ–∑–ª–µ —Å—Ç—Ä–æ–∫ —á—Ç–æ –æ–Ω–∏ –¥–µ–ª–∞—é—Ç –∏ –¥–ª—è —á–µ–≥–æ
–ü—Ä–æ—Å—Ç–∞—è CNN
–£–ª—É—á—à–µ–Ω–Ω–∞—è CNN
EfficientNet-like
ResNet-like
üèóÔ∏è –ü–æ–¥—Ä–æ–±–Ω–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä —Å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏—è–º–∏
1. üéØ –ü–†–û–°–¢–ê–Ø CNN –ê–†–•–ò–¢–ï–ö–¢–£–†–ê
python
def create_simple_cnn(input_shape=(224, 224, 3), num_classes=8):
    """
    –ü–†–û–°–¢–ê–Ø CNN - –∏–¥–µ–∞–ª—å–Ω–∞ –¥–ª—è –±—ã—Å—Ç—Ä–æ–≥–æ —Å—Ç–∞—Ä—Ç–∞ –∏ –±–∞–∑–æ–≤—ã—Ö –∑–∞–¥–∞—á
    –õ–∏–Ω–µ–π–Ω–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞ –±–µ–∑ —Å–ª–æ–∂–Ω—ã—Ö —Å–æ–µ–¥–∏–Ω–µ–Ω–∏–π
    """
    model = Sequential([
        # === –ë–õ–û–ö 1: –ù–ê–ß–ê–õ–¨–ù–ê–Ø –û–ë–†–ê–ë–û–¢–ö–ê –ò–ó–û–ë–†–ê–ñ–ï–ù–ò–Ø ===
        # –ü–µ—Ä–≤–∞—è —Å–≤–µ—Ä—Ç–∫–∞ - –∏–∑–≤–ª–µ–∫–∞–µ—Ç –±–∞–∑–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ (–∫—Ä–∞—è, —Ç–µ–∫—Å—Ç—É—Ä—ã)
        layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape),
        # –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è - —É—Å–∫–æ—Ä—è–µ—Ç –æ–±—É—á–µ–Ω–∏–µ, —Å—Ç–∞–±–∏–ª–∏–∑–∏—Ä—É–µ—Ç –≥—Ä–∞–¥–∏–µ–Ω—Ç—ã
        layers.BatchNormalization(),
        # –í—Ç–æ—Ä–∞—è —Å–≤–µ—Ä—Ç–∫–∞ - —É–≥–ª—É–±–ª—è–µ—Ç –∏–∑–≤–ª–µ—á–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        # –ü—É–ª–ª–∏–Ω–≥ - —É–º–µ–Ω—å—à–∞–µ—Ç —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å, —Å–æ—Ö—Ä–∞–Ω—è—è –≤–∞–∂–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
        layers.MaxPooling2D(2),
        # Dropout - –ø—Ä–µ–¥–æ—Ç–≤—Ä–∞—â–∞–µ—Ç –∑–∞–ø–æ–º–∏–Ω–∞–Ω–∏–µ —à—É–º–∞ (–±–æ—Ä—å–±–∞ —Å –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ–º)
        layers.Dropout(0.25),
        
        # === –ë–õ–û–ö 2: –ü–†–û–ú–ï–ñ–£–¢–û–ß–ù–´–ï –ü–†–ò–ó–ù–ê–ö–ò ===
        # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Ñ–∏–ª—å—Ç—Ä—ã –¥–ª—è –±–æ–ª–µ–µ —Å–ª–æ–∂–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(2),
        layers.Dropout(0.25),
        
        # === –ë–õ–û–ö 3: –°–õ–û–ñ–ù–´–ï –ü–†–ò–ó–ù–ê–ö–ò ===
        # –ï—â–µ –±–æ–ª—å—à–µ —Ñ–∏–ª—å—Ç—Ä–æ–≤ –¥–ª—è –≤—ã—Å–æ–∫–æ—É—Ä–æ–≤–Ω–µ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(2),
        layers.Dropout(0.25),
        
        # === –ë–õ–û–ö 4: –§–ò–ù–ê–õ–¨–ù–´–ï –ü–†–ò–ó–ù–ê–ö–ò ===
        # –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ñ–∏–ª—å—Ç—Ä–æ–≤ –¥–ª—è —Å–∞–º—ã—Ö –∞–±—Å—Ç—Ä–∞–∫—Ç–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
        layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        # –ì–ª–æ–±–∞–ª—å–Ω—ã–π –ø—É–ª–ª–∏–Ω–≥ - –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç 2D –∫–∞—Ä—Ç—ã –≤ 1D –≤–µ–∫—Ç–æ—Ä
        layers.GlobalAveragePooling2D(),
        # –°–∏–ª—å–Ω—ã–π dropout –ø–µ—Ä–µ–¥ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–µ–π
        layers.Dropout(0.5),
        
        # === –ö–õ–ê–°–°–ò–§–ò–ö–ê–¢–û–† ===
        # –ü–æ–ª–Ω–æ—Å–≤—è–∑–Ω—ã–π —Å–ª–æ–π - learns non-linear combinations
        layers.Dense(512, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        # –£–º–µ–Ω—å—à–∞–µ–º —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –ø–µ—Ä–µ–¥ –≤—ã—Ö–æ–¥–æ–º
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.2),
        
        # –í–´–•–û–î - –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è 8 –∫–ª–∞—Å—Å–æ–≤
        layers.Dense(num_classes, activation='softmax')
    ])
    
    return model
üéØ –î–ª—è —á–µ–≥–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å: –ë—ã—Å—Ç—Ä–æ–µ –ø—Ä–æ—Ç–æ—Ç–∏–ø–∏—Ä–æ–≤–∞–Ω–∏–µ, –º–∞–ª–µ–Ω—å–∫–∏–µ –¥–∞—Ç–∞—Å–µ—Ç—ã, –±–∞–∑–æ–≤—ã–µ –∑–∞–¥–∞—á–∏

2. üöÄ –£–õ–£–ß–®–ï–ù–ù–ê–Ø CNN –° RESIDUAL –°–í–Ø–ó–Ø–ú–ò
python
def create_improved_cnn(input_shape=(224, 224, 3), num_classes=8):
    """
    –£–õ–£–ß–®–ï–ù–ù–ê–Ø CNN - –¥–æ–±–∞–≤–ª—è–µ—Ç skip-connections –¥–ª—è –±–æ—Ä—å–±—ã —Å vanishing gradient
    –ü–æ–∑–≤–æ–ª—è–µ—Ç —Å—Ç—Ä–æ–∏—Ç—å –±–æ–ª–µ–µ –≥–ª—É–±–æ–∫–∏–µ —Å–µ—Ç–∏
    """
    inputs = tf.keras.Input(shape=input_shape)
    
    # === STEM –ë–õ–û–ö: –ü–ï–†–í–ò–ß–ù–ê–Ø –û–ë–†–ê–ë–û–¢–ö–ê ===
    # –ë–æ–ª—å—à–æ–µ —è–¥—Ä–æ 7x7 - –∑–∞—Ö–≤–∞—Ç—ã–≤–∞–µ—Ç –±–æ–ª—å—à–∏–µ –ø–∞—Ç—Ç–µ—Ä–Ω—ã
    x = layers.Conv2D(32, (7, 7), strides=2, padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    # –ê–≥—Ä–µ—Å—Å–∏–≤–Ω—ã–π –ø—É–ª–ª–∏–Ω–≥ - –±—ã—Å—Ç—Ä–æ —É–º–µ–Ω—å—à–∞–µ—Ç —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å
    x = layers.MaxPooling2D(3, strides=2, padding='same')(x)
    
    # === RESIDUAL BLOCK 1: –ë–ê–ó–û–í–´–ï –ü–†–ò–ó–ù–ê–ö–ò ===
    shortcut = x  # –°–æ—Ö—Ä–∞–Ω—è–µ–º –≤—Ö–æ–¥ –¥–ª—è skip-connection
    
    # –î–≤–µ —Å–≤–µ—Ä—Ç–∫–∏ –¥–ª—è –∏–∑–≤–ª–µ—á–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    x = layers.Conv2D(64, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(64, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    
    # PROJECTION: –µ—Å–ª–∏ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –Ω–µ —Å–æ–≤–ø–∞–¥–∞—é—Ç, –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º shortcut
    if shortcut.shape[-1] != 64:
        shortcut = layers.Conv2D(64, (1, 1))(shortcut)  # 1x1 conv –º–µ–Ω—è–µ—Ç channels
        shortcut = layers.BatchNormalization()(shortcut)
    
    # SKIP-CONNECTION: –¥–æ–±–∞–≤–ª—è–µ–º –∏—Å—Ö–æ–¥–Ω—ã–π —Å–∏–≥–Ω–∞–ª (—Ä–µ—à–∞–µ—Ç vanishing gradient)
    x = layers.Add()([x, shortcut])
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(2)(x)
    
    # === RESIDUAL BLOCK 2: –°–õ–û–ñ–ù–´–ï –ü–†–ò–ó–ù–ê–ö–ò ===
    shortcut = x
    x = layers.Conv2D(128, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(128, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    
    # –ü—Ä—è–º–æ–π skip-connection (—Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ —Å–æ–≤–ø–∞–¥–∞—é—Ç)
    x = layers.Add()([x, shortcut])
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(2)(x)
    
    # === RESIDUAL BLOCK 3: –í–´–°–û–ö–û–£–†–û–í–ù–ï–í–´–ï –ü–†–ò–ó–ù–ê–ö–ò ===
    shortcut = x
    x = layers.Conv2D(256, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(256, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    
    x = layers.Add()([x, shortcut])
    x = layers.Activation('relu')(x)
    
    # === –ö–õ–ê–°–°–ò–§–ò–ö–ê–¢–û–† ===
    x = layers.GlobalAveragePooling2D()(x)  # –ë–æ–ª–µ–µ —Å—Ç–∞–±–∏–ª—å–Ω–æ —á–µ–º Flatten()
    x = layers.Dense(512, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    
    model = tf.keras.Model(inputs, outputs)
    return model
üéØ –î–ª—è —á–µ–≥–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å: –ì–ª—É–±–æ–∫–∏–µ —Å–µ—Ç–∏, —Å–ª–æ–∂–Ω—ã–µ –¥–∞—Ç–∞—Å–µ—Ç—ã, –∫–æ–≥–¥–∞ –ø—Ä–æ—Å—Ç–∞—è CNN –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–∞

3. ‚ö° EFFICIENTNET-LIKE –ê–†–•–ò–¢–ï–ö–¢–£–†–ê
python
def create_efficientnet_like(input_shape=(224, 224, 3), num_classes=8):
    """
    EFFICIENTNET-LIKE - —Å–æ–≤—Ä–µ–º–µ–Ω–Ω–∞—è —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞
    –ò—Å–ø–æ–ª—å–∑—É–µ—Ç depthwise convolutions –∏ squeeze-excitation
    –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –ø—Ä–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–∞—Ö
    """
    def mb_conv(x, filters, stride=1):
        """
        MBConv –±–ª–æ–∫ - –æ—Å–Ω–æ–≤–∞ EfficientNet
        –°–æ—á–µ—Ç–∞–µ—Ç depthwise conv –∏ squeeze-excitation
        """
        shortcut = x  # –î–ª—è residual connection
        
        # === DEPTHWISE CONVOLUTION ===
        # –û—Ç–¥–µ–ª—å–Ω–∞—è —Å–≤–µ—Ä—Ç–∫–∞ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–∞–Ω–∞–ª–∞ (—ç–∫–æ–Ω–æ–º–Ω–æ –ø–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º)
        x = layers.DepthwiseConv2D((3, 3), strides=stride, padding='same')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation('swish')(x)  # Swish - —Å–æ–≤—Ä–µ–º–µ–Ω–Ω–∞—è –∑–∞–º–µ–Ω–∞ ReLU
        
        # === SQUEEZE-AND-EXCITATION ===
        # –ú–µ—Ö–∞–Ω–∏–∑–º –≤–Ω–∏–º–∞–Ω–∏—è –∫ –∫–∞–Ω–∞–ª–∞–º
        se = layers.GlobalAveragePooling2D()(x)  # –°–∂–∏–º–∞–µ–º –¥–æ –≤–µ–∫—Ç–æ—Ä–∞
        se = layers.Reshape((1, 1, filters))(se)  # –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–ª—è broadcast
        # –î–≤–∞ –ø–æ–ª–Ω–æ—Å–≤—è–∑–Ω—ã—Ö —Å–ª–æ—è –¥–ª—è –ø–µ—Ä–µ–≤–∑–≤–µ—à–∏–≤–∞–Ω–∏—è –∫–∞–Ω–∞–ª–æ–≤
        se = layers.Dense(filters // 4, activation='swish')(se)  # –°–∂–∞—Ç–∏–µ
        se = layers.Dense(filters, activation='sigmoid')(se)     # –í–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏–µ
        # –£–º–Ω–æ–∂–∞–µ–º –∫–∞–∂–¥—ã–π –∫–∞–Ω–∞–ª –Ω–∞ –µ–≥–æ "–≤–∞–∂–Ω–æ—Å—Ç—å"
        x = layers.Multiply()([x, se])
        
        # === POINTWISE CONVOLUTION ===
        # 1x1 —Å–≤–µ—Ä—Ç–∫–∞ –∫–æ–º–±–∏–Ω–∏—Ä—É–µ—Ç –∫–∞–Ω–∞–ª—ã
        x = layers.Conv2D(filters, (1, 1), padding='same')(x)
        x = layers.BatchNormalization()(x)
        
        # === RESIDUAL CONNECTION ===
        # –î–æ–±–∞–≤–ª—è–µ–º shortcut –µ—Å–ª–∏ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ —Å–æ–≤–ø–∞–¥–∞—é—Ç
        if stride == 1 and shortcut.shape[-1] == filters:
            x = layers.Add()([x, shortcut])
        
        return x
    
    inputs = tf.keras.Input(shape=input_shape)
    
    # === STEM –ë–õ–û–ö ===
    # –ù–∞—á–∞–ª—å–Ω–∞—è —Å–≤–µ—Ä—Ç–∫–∞ —Å –±–æ–ª—å—à–∏–º —à–∞–≥–æ–º –¥–ª—è –±—ã—Å—Ç—Ä–æ–≥–æ —É–º–µ–Ω—å—à–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–∞
    x = layers.Conv2D(32, (3, 3), strides=2, padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('swish')(x)
    
    # === –ü–û–°–õ–ï–î–û–í–ê–¢–ï–õ–¨–ù–û–°–¢–¨ MBConv –ë–õ–û–ö–û–í ===
    # –ü–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ–µ —É–≤–µ–ª–∏—á–µ–Ω–∏–µ —Ñ–∏–ª—å—Ç—Ä–æ–≤ –∏ —É–º–µ–Ω—å—à–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–∞
    x = mb_conv(x, 16, stride=1)   # –ù–∞—á–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
    x = mb_conv(x, 24, stride=2)   # –£–º–µ–Ω—å—à–∞–µ–º —Ä–∞–∑–º–µ—Ä –≤ 2 —Ä–∞–∑–∞
    x = mb_conv(x, 40, stride=2)   # –ï—â–µ —É–º–µ–Ω—å—à–∞–µ–º
    x = mb_conv(x, 80, stride=2)   # –ü—Ä–æ–¥–æ–ª–∂–∞–µ–º —É–≥–ª—É–±–ª—è—Ç—å
    x = mb_conv(x, 112, stride=1)  # –ë–µ–∑ —É–º–µ–Ω—å—à–µ–Ω–∏—è, —Ç–æ–ª—å–∫–æ –ø—Ä–∏–∑–Ω–∞–∫–∏
    x = mb_conv(x, 192, stride=2)  # –§–∏–Ω–∞–ª—å–Ω–æ–µ —É–º–µ–Ω—å—à–µ–Ω–∏–µ
    x = mb_conv(x, 320, stride=1)  # –§–∏–Ω–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
    
    # === HEAD –ë–õ–û–ö ===
    # –§–∏–Ω–∞–ª—å–Ω–∞—è 1x1 —Å–≤–µ—Ä—Ç–∫–∞ –¥–ª—è —É–≤–µ–ª–∏—á–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    x = layers.Conv2D(1280, (1, 1), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('swish')(x)
    # –ì–ª–æ–±–∞–ª—å–Ω—ã–π –ø—É–ª–ª–∏–Ω–≥ –¥–ª—è –ø–µ—Ä–µ—Ö–æ–¥–∞ –∫ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.3)(x)
    # –í—ã—Ö–æ–¥–Ω–æ–π —Å–ª–æ–π
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    
    model = tf.keras.Model(inputs, outputs)
    return model
üéØ –î–ª—è —á–µ–≥–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å: –ú–æ–±–∏–ª—å–Ω—ã–µ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è, –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω—ã–µ —Ä–µ—Å—É—Ä—Å—ã, –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å

4. üèÜ RESNET-LIKE –ê–†–•–ò–¢–ï–ö–¢–£–†–ê
python
def create_resnet_like(input_shape=(224, 224, 3), num_classes=8):
    """
    RESNET-LIKE - –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞ —Å –≥–ª—É–±–æ–∫–∏–º–∏ residual –±–ª–æ–∫–∞–º–∏
    –†–µ—à–∞–µ—Ç –ø—Ä–æ–±–ª–µ–º—É –∏—Å—á–µ–∑–∞—é—â–∏—Ö –≥—Ä–∞–¥–∏–µ–Ω—Ç–æ–≤ –≤ –æ—á–µ–Ω—å –≥–ª—É–±–æ–∫–∏—Ö —Å–µ—Ç—è—Ö
    """
    def residual_block(x, filters, downsample=False):
        """
        Residual –±–ª–æ–∫ - –æ—Å–Ω–æ–≤–∞ ResNet
        –í—Å–µ–≥–¥–∞ —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç –≥—Ä–∞–¥–∏–µ–Ω—Ç —á–µ—Ä–µ–∑ skip-connections
        """
        shortcut = x
        
        # === –ü–ï–†–í–ê–Ø –°–í–ï–†–¢–ö–ê –ë–õ–û–ö–ê ===
        x = layers.Conv2D(filters, (3, 3), 
                         strides=(2 if downsample else 1),  # –£–º–µ–Ω—å—à–∞–µ–º –µ—Å–ª–∏ –Ω—É–∂–Ω–æ
                         padding='same')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation('relu')(x)
        
        # === –í–¢–û–†–ê–Ø –°–í–ï–†–¢–ö–ê –ë–õ–û–ö–ê ===
        x = layers.Conv2D(filters, (3, 3), padding='same')(x)
        x = layers.BatchNormalization()(x)
        
        # === SKIP-CONNECTION –û–ë–†–ê–ë–û–¢–ö–ê ===
        if downsample or shortcut.shape[-1] != filters:
            # PROJECTION SHORTCUT: –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º shortcut —á—Ç–æ–±—ã —Å–æ–≤–ø–∞–ª–∏ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏
            shortcut = layers.Conv2D(filters, (1, 1), 
                                   strides=(2 if downsample else 1))(shortcut)
            shortcut = layers.BatchNormalization()(shortcut)
        
        # === ADD AND ACTIVATE ===
        # –°–∫–ª–∞–¥—ã–≤–∞–µ–º –æ—Å–Ω–æ–≤–Ω–æ–π –ø—É—Ç—å –∏ shortcut
        x = layers.Add()([x, shortcut])
        # –ê–∫—Ç–∏–≤–∞—Ü–∏—è –ü–û–°–õ–ï —Å–ª–æ–∂–µ–Ω–∏—è (ResNet v1)
        x = layers.Activation('relu')(x)
        
        return x
    
    inputs = tf.keras.Input(shape=input_shape)
    
    # === INITIAL CONV ===
    # –ë–æ–ª—å—à–æ–µ —è–¥—Ä–æ –¥–ª—è –∑–∞—Ö–≤–∞—Ç–∞ –∫—Ä—É–ø–Ω—ã—Ö –ø–∞—Ç—Ç–µ—Ä–Ω–æ–≤
    x = layers.Conv2D(64, (7, 7), strides=2, padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(3, strides=2, padding='same')(x)
    
    # === RESIDUAL BLOCKS ===
    # Conv2_x (3 –±–ª–æ–∫–∞)
    for _ in range(2):
        x = residual_block(x, 64)
    
    # Conv3_x (4 –±–ª–æ–∫–∞) + downsample
    x = residual_block(x, 128, downsample=True)
    for _ in range(3):
        x = residual_block(x, 128)
    
    # Conv4_x (6 –±–ª–æ–∫–æ–≤) + downsample
    x = residual_block(x, 256, downsample=True)
    for _ in range(5):
        x = residual_block(x, 256)
    
    # Conv5_x (3 –±–ª–æ–∫–∞) + downsample
    x = residual_block(x, 512, downsample=True)
    for _ in range(2):
        x = residual_block(x, 512)
    
    # === CLASSIFICATION HEAD ===
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(512, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    
    model = tf.keras.Model(inputs, outputs)
    return model
üéØ –î–ª—è —á–µ–≥–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å: –û—á–µ–Ω—å –≥–ª—É–±–æ–∫–∏–µ —Å–µ—Ç–∏, –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å, –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ –∑–∞–¥–∞—á–∏

üìä –°–í–û–î–ù–ê–Ø –¢–ê–ë–õ–ò–¶–ê –í–´–ë–û–†–ê:
–ê—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞	–ì–ª—É–±–∏–Ω–∞	–ü–∞—Ä–∞–º–µ—Ç—Ä—ã	–°–∫–æ—Ä–æ—Å—Ç—å	–¢–æ—á–Ω–æ—Å—Ç—å	–ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ
–ü—Ä–æ—Å—Ç–∞—è CNN	10-15 —Å–ª–æ–µ–≤	~1-2M	üöÄ –ë—ã—Å—Ç—Ä–∞—è	85-90%	–ë–∞–∑–æ–≤—ã–µ –∑–∞–¥–∞—á–∏, —Å—Ç–∞—Ä—Ç
–£–ª—É—á—à–µ–Ω–Ω–∞—è CNN	15-20 —Å–ª–æ–µ–≤	~3-5M	‚ö° –°—Ä–µ–¥–Ω—è—è	88-92%	–ü—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–æ, –±–∞–ª–∞–Ω—Å
EfficientNet-like	20-30 —Å–ª–æ–µ–≤	~2-3M	üê¢ –ú–µ–¥–ª–µ–Ω–Ω–∞—è	90-94%	–ú–æ–±–∏–ª—å–Ω—ã–µ, —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å
ResNet-like	50+ —Å–ª–æ–µ–≤	~10-15M	üêå –û—á–µ–Ω—å –º–µ–¥–ª.	92-96%	–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å
üèÜ –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—è: –ù–∞—á–Ω–∏—Ç–µ —Å –ü—Ä–æ—Å—Ç–æ–π CNN, –∑–∞—Ç–µ–º –ø–µ—Ä–µ—Ö–æ–¥–∏—Ç–µ –∫ –£–ª—É—á—à–µ–Ω–Ω–æ–π CNN –ø—Ä–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏! üöÄ
```

```python
–ë–ª–æ–∫–∏ –≤ AutoModel
- –Ω–∞–ø–∏—à–∏ –∫–∞–∫–∏–µ –±–ª–æ–∫–∏ —Å—É—â–µ—Å—Ç–≤—É—é—Ç,
- –∑–∞ —á—Ç–æ –æ—Ç–≤–µ—á–∞—é—Ç,
- –∫–∞–∫–æ–≤–∞ —Ç–∏–ø–æ–≤–∞—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –¥–ª—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π (–ø—Ä–µ–¥–ª–æ–∂–∏ —Ç—Ä–∏ –≤–∞—Ä–∏–∞–Ω—Ç–∞: –ø—Ä–æ—Å—Ç–æ–π, —Å—Ä–µ–¥–Ω–∏–π –∏ —Å–ª–æ–∂–Ω—ã–π)

–∫–∞–∫–æ–π –ø–æ–¥–æ–π–¥–µ—Ç –∑–∞–≥—Ä—É–∑—á–∏–∫
üèóÔ∏è –ë–ª–æ–∫–∏ –≤ AutoModel –¥–ª—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
**1. üì• –í—Ö–æ–¥–Ω—ã–µ –±–ª–æ–∫–∏
python
# –î–ª—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
input_node = ak.ImageInput()
# –î–ª—è —Ç–∞–±–ª–∏—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö  
# input_node = ak.StructuredDataInput()
# –î–ª—è —Ç–µ–∫—Å—Ç–∞
# input_node = ak.TextInput()
**2. üé® –ë–ª–æ–∫–∏ –∞—É–≥–º–µ–Ω—Ç–∞—Ü–∏–∏
python
augmentation = ak.ImageAugmentation(
    horizontal_flip=True,      # –ì–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–µ –æ—Ç—Ä–∞–∂–µ–Ω–∏–µ
    vertical_flip=False,       # –í–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–µ –æ—Ç—Ä–∞–∂–µ–Ω–∏–µ (—Ä–µ–¥–∫–æ –¥–ª—è —Ñ–æ—Ç–æ)
    rotation_factor=0.1,       # –ü–æ–≤–æ—Ä–æ—Ç ¬±10%
    zoom_factor=0.2,           # –ó—É–º ¬±20%
    contrast_factor=0.2,       # –ö–æ–Ω—Ç—Ä–∞—Å—Ç ¬±20%
    translation_factor=0.1     # –°–¥–≤–∏–≥ ¬±10%
)
**3. üîç –°–≤–µ—Ä—Ç–æ—á–Ω—ã–µ –±–ª–æ–∫–∏
python
# –ë–∞–∑–æ–≤—ã–π —Å–≤–µ—Ä—Ç–æ—á–Ω—ã–π –±–ª–æ–∫
conv_block = ak.ConvBlock(
    num_blocks=2,           # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–ª–æ–∫–æ–≤ (1-4)
    num_layers=2,           # –°–ª–æ–µ–≤ –≤ –±–ª–æ–∫–µ (1-3)
    filters=32,             # –ù–∞—á–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ñ–∏–ª—å—Ç—Ä–æ–≤
    kernel_size=3,          # –†–∞–∑–º–µ—Ä —è–¥—Ä–∞ (3, 5, 7)
    max_pooling=True,       # –î–æ–±–∞–≤–ª—è—Ç—å MaxPooling –ø–æ—Å–ª–µ –±–ª–æ–∫–æ–≤
    separable=False,        # –ò—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å DepthwiseConv2D
    dropout=0.25            # Dropout –º–µ–∂–¥—É —Å–ª–æ—è–º–∏
)

# ResNet –±–ª–æ–∫ (—Å skip-connections)
resnet_block = ak.ResNetBlock(
    version='v2',           # v1 –∏–ª–∏ v2 –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞
    pooling='avg'           # –¢–∏–ø –ø—É–ª–ª–∏–Ω–≥–∞
)

# Xception –±–ª–æ–∫ (–≥–ª—É–±–æ–∫–∏–µ separable —Å–≤–µ—Ä—Ç–∫–∏)
xception_block = ak.XceptionBlock(
    num_blocks=2,
    dropout=0.2
)
**4. üìê –ë–ª–æ–∫–∏ –ø—É–ª–ª–∏–Ω–≥–∞ –∏ –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏–∏
python
# –ü—É–ª–ª–∏–Ω–≥
max_pool = ak.MaxPooling2D(pool_size=2)
avg_pool = ak.AveragePooling2D(pool_size=2)
global_avg_pool = ak.GlobalAveragePooling2D()
global_max_pool = ak.GlobalMaxPooling2D()

# –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è
normalization = ak.Normalization()  # –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è
batch_norm = ak.BatchNormalization()
**5. üß† –ë–ª–æ–∫–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏/—Ä–µ–≥—Ä–µ—Å—Å–∏–∏
python
# –î–ª—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
classification_head = ak.ClassificationHead(
    num_classes=8,
    dropout=0.5,
    metrics=['accuracy']
)

# –î–ª—è —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
regression_head = ak.RegressionHead(
    loss='mean_squared_error',
    metrics=['mae']
)

# –ü–æ–ª–Ω–æ—Å–≤—è–∑–Ω—ã–µ —Å–ª–æ–∏
dense_block = ak.DenseBlock(
    num_layers=2,
    dropout=0.5,
    use_batchnorm=True
)
**6. üîÑ –ë–ª–æ–∫–∏ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è
python
# –î–ª—è –º–Ω–æ–≥–æ–º–æ–¥–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
merged = ak.Merge()([branch1, branch2, branch3])

# –î–ª—è skip-connections (–≤—Ä—É—á–Ω—É—é)
# merged = layers.Add()([x, shortcut])
üöÄ –¢–∏–ø–æ–≤—ã–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –¥–ª—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
–í–∞—Ä–∏–∞–Ω—Ç 1: üü¢ –ü–†–û–°–¢–û–ô (–±—ã—Å—Ç—Ä—ã–π —Å—Ç–∞—Ä—Ç)
python
import autokeras as ak

def create_simple_pipeline():
    """–ü—Ä–æ—Å—Ç–æ–π –ø–∞–π–ø–ª–∞–π–Ω - –∏–¥–µ–∞–ª–µ–Ω –¥–ª—è –±—ã—Å—Ç—Ä–æ–≥–æ —Å—Ç–∞—Ä—Ç–∞"""
    input_node = ak.ImageInput()
    
    # –ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è –∞—É–≥–º–µ–Ω—Ç–∞—Ü–∏—è
    augmented = ak.ImageAugmentation(horizontal_flip=True)(input_node)
    
    # –ë–∞–∑–æ–≤—ã–µ —Å–≤–µ—Ä—Ç–∫–∏
    features = ak.ConvBlock(
        num_blocks=2,
        num_layers=2,
        max_pooling=True,
        dropout=0.2
    )(augmented)
    
    # –ü—Ä–æ—Å—Ç–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è
    output = ak.ClassificationHead(num_classes=8)(features)
    
    return ak.AutoModel(
        inputs=input_node,
        outputs=output,
        max_trials=10,
        overwrite=True
    )

# üéØ –î–ª—è: –ë—ã—Å—Ç—Ä–æ–≥–æ –ø—Ä–æ—Ç–æ—Ç–∏–ø–∏—Ä–æ–≤–∞–Ω–∏—è, –º–∞–ª–µ–Ω—å–∫–∏—Ö –¥–∞—Ç–∞—Å–µ—Ç–æ–≤
–í–∞—Ä–∏–∞–Ω—Ç 2: üü° –°–†–ï–î–ù–ò–ô (–±–∞–ª–∞–Ω—Å —Å–∫–æ—Ä–æ—Å—Ç–∏ –∏ –∫–∞—á–µ—Å—Ç–≤–∞)
python
def create_balanced_pipeline():
    """–°–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–∞–π–ø–ª–∞–π–Ω - –ª—É—á—à–∏–π –≤—ã–±–æ—Ä –¥–ª—è –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–∞ –∑–∞–¥–∞—á"""
    input_node = ak.ImageInput()
    
    # –†–∞—Å—à–∏—Ä–µ–Ω–Ω–∞—è –∞—É–≥–º–µ–Ω—Ç–∞—Ü–∏—è
    augmented = ak.ImageAugmentation(
        horizontal_flip=True,
        rotation_factor=0.1,
        zoom_factor=0.1,
        contrast_factor=0.1
    )(input_node)
    
    # –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è
    normalized = ak.Normalization()(augmented)
    
    # ResNet-like –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞
    features = ak.ResNetBlock(version='v2')(normalized)
    
    # –ì–ª–æ–±–∞–ª—å–Ω—ã–π –ø—É–ª–ª–∏–Ω–≥
    pooled = ak.GlobalAveragePooling2D()(features)
    
    # Dense –±–ª–æ–∫ –ø–µ—Ä–µ–¥ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–µ–π
    dense = ak.DenseBlock(num_layers=1, dropout=0.3)(pooled)
    
    output = ak.ClassificationHead(num_classes=8)(dense)
    
    return ak.AutoModel(
        inputs=input_node,
        outputs=output,
        max_trials=15,
        overwrite=True
    )

# üéØ –î–ª—è: –ü—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∑–∞–¥–∞—á, —Å—Ä–µ–¥–Ω–∏—Ö –¥–∞—Ç–∞—Å–µ—Ç–æ–≤
–í–∞—Ä–∏–∞–Ω—Ç 3: üî¥ –°–õ–û–ñ–ù–´–ô (–º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å)
python
def create_advanced_pipeline():
    """–ü—Ä–æ–¥–≤–∏–Ω—É—Ç—ã–π –ø–∞–π–ø–ª–∞–π–Ω - –¥–ª—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π —Ç–æ—á–Ω–æ—Å—Ç–∏"""
    input_node = ak.ImageInput()
    
    # –ê–≥—Ä–µ—Å—Å–∏–≤–Ω–∞—è –∞—É–≥–º–µ–Ω—Ç–∞—Ü–∏—è
    augmented = ak.ImageAugmentation(
        horizontal_flip=True,
        vertical_flip=True,
        rotation_factor=0.2,
        zoom_factor=0.2,
        contrast_factor=0.2,
        translation_factor=0.1
    )(input_node)
    
    # –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è
    normalized = ak.Normalization()(augmented)
    
    # –ú—É–ª—å—Ç–∏-–∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–Ω—ã–π –ø–æ–¥—Ö–æ–¥
    # –í–µ—Ç–∫–∞ 1: ResNet
    branch1 = ak.ResNetBlock(version='v2')(normalized)
    branch1 = ak.GlobalAveragePooling2D()(branch1)
    
    # –í–µ—Ç–∫–∞ 2: Xception
    branch2 = ak.XceptionBlock(num_blocks=2)(normalized)
    branch2 = ak.GlobalMaxPooling2D()(branch2)
    
    # –í–µ—Ç–∫–∞ 3: Custom CNN
    branch3 = ak.ConvBlock(
        num_blocks=3,
        num_layers=2,
        separable=True,
        dropout=0.3
    )(normalized)
    branch3 = ak.GlobalAveragePooling2D()(branch3)
    
    # –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –≤–µ—Ç–æ–∫
    merged = ak.Merge()([branch1, branch2, branch3])
    
    # –†–∞—Å—à–∏—Ä–µ–Ω–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è
    dense = ak.DenseBlock(
        num_layers=2,
        dropout=0.5,
        use_batchnorm=True
    )(merged)
    
    output = ak.ClassificationHead(num_classes=8)(dense)
    
    return ak.AutoModel(
        inputs=input_node,
        outputs=output,
        max_trials=20,
        overwrite=True
    )

# üéØ –î–ª—è: –°–æ—Ä–µ–≤–Ω–æ–≤–∞–Ω–∏–π, –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–π, –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π —Ç–æ—á–Ω–æ—Å—Ç–∏
üì• –ü–æ–¥—Ö–æ–¥—è—â–∏–π –∑–∞–≥—Ä—É–∑—á–∏–∫ –¥–∞–Ω–Ω—ã—Ö
–†–µ–∫–æ–º–µ–Ω–¥—É–µ–º—ã–π –∑–∞–≥—Ä—É–∑—á–∏–∫ –¥–ª—è AutoKeras:
python
import tensorflow as tf

def create_optimal_loader(data_dir, batch_size=32, img_size=(224, 224)):
    """
    –û–ø—Ç–∏–º–∞–ª—å–Ω—ã–π –∑–∞–≥—Ä—É–∑—á–∏–∫ –¥–ª—è AutoKeras
    AutoKeras —Å–∞–º –¥–µ–ª–∞–µ—Ç –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—é –∏ —á–∞—Å—Ç—å –∞—É–≥–º–µ–Ω—Ç–∞—Ü–∏–∏!
    """
    
    # –°—ã—Ä—ã–µ –¥–∞–Ω–Ω—ã–µ –ë–ï–ó –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏
    train_data = tf.keras.preprocessing.image_dataset_from_directory(
        data_dir,
        validation_split=0.2,
        subset="training",
        seed=42,
        image_size=img_size,
        batch_size=batch_size,
        shuffle=True,
        label_mode='categorical'  # –î–ª—è ClassificationHead
    )
    
    validation_data = tf.keras.preprocessing.image_dataset_from_directory(
        data_dir,
        validation_split=0.2,
        subset="validation",
        seed=42,
        image_size=img_size,
        batch_size=batch_size,
        shuffle=False,  # –î–ª—è –≤–∞–ª–∏–¥–∞—Ü–∏–∏ –Ω–µ –Ω—É–∂–Ω–æ shuffle
        label_mode='categorical'
    )
    
    # –¢–æ–ª—å–∫–æ –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏
    # AutoKeras —Å–∞–º —Å–¥–µ–ª–∞–µ—Ç –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—é –∏ –∞—É–≥–º–µ–Ω—Ç–∞—Ü–∏—é!
    train_data = train_data.prefetch(tf.data.AUTOTUNE)
    validation_data = validation_data.prefetch(tf.data.AUTOTUNE)
    
    return train_data, validation_data

# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ
train_data, val_data = create_optimal_loader('path/to/your/images')
–ß–µ–≥–æ –ù–ï –¥–µ–ª–∞—Ç—å –≤ –∑–∞–≥—Ä—É–∑—á–∏–∫–µ –¥–ª—è AutoKeras:
python
# ‚ùå –ù–ï –î–ï–õ–ê–¢–¨ - AutoKeras —Å–¥–µ–ª–∞–µ—Ç —ç—Ç–æ —Å–∞–º:
# - –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—é –ø–∏–∫—Å–µ–ª–µ–π
# - –ê—É–≥–º–µ–Ω—Ç–∞—Ü–∏—é –¥–∞–Ω–Ω—ã—Ö  
# - Rescaling
# - One-hot encoding

# ‚úÖ –î–ï–õ–ê–¢–¨:
# - –ó–∞–≥—Ä—É–∑–∫—É –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
# - –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ train/val
# - Prefetch –¥–ª—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏
üéØ –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏ –ø–æ –≤—ã–±–æ—Ä—É:
–ö—Ä–∏—Ç–µ—Ä–∏–π	–ü—Ä–æ—Å—Ç–æ–π	–°—Ä–µ–¥–Ω–∏–π	–°–ª–æ–∂–Ω—ã–π
–í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è	1-2 —á–∞—Å–∞	3-6 —á–∞—Å–æ–≤	8+ —á–∞—Å–æ–≤
–¢–æ—á–Ω–æ—Å—Ç—å	85-90%	90-95%	95-98%
–í—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω—ã–µ —Ä–µ—Å—É—Ä—Å—ã	–ù–∏–∑–∫–∏–µ	–°—Ä–µ–¥–Ω–∏–µ	–í—ã—Å–æ–∫–∏–µ
–†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—è	üèÜ –ù–∞—á–∏–Ω–∞—Ç—å –∑–¥–µ—Å—å!	–î–ª—è –ø—Ä–æ–¥–∞–∫—à–µ–Ω–∞	–î–ª—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π —Ç–æ—á–Ω–æ—Å—Ç–∏
–°–æ–≤–µ—Ç: –ù–∞—á–Ω–∏—Ç–µ —Å –ü–†–û–°–¢–û–ì–û –≤–∞—Ä–∏–∞–Ω—Ç–∞, –∑–∞—Ç–µ–º –ø–µ—Ä–µ—Ö–æ–¥–∏—Ç–µ –∫ –°–†–ï–î–ù–ï–ú–£ –µ—Å–ª–∏ –Ω—É–∂–Ω–∞ –ª—É—á—à–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å! üöÄ
```
